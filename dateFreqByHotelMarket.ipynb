```python
def dd():
    return defaultdict(int)

checkInFreqByHotelMarket = defaultdict(dd)
checkOutFreqByHotelMarket = defaultdict(dd)
```

```python
print('Opening train...')
f = open("data/train.csv", "r")
f.readline()
total = 0
    
while 1:
    line = f.readline().strip()
    total += 1
        
    if total % 1000000 == 0:
        print('Read {} lines...'.format(total))
    if line == '':
         break

    arr = line.split(",")
    checkInDate = arr[11]
    checkOutDate = arr[12]
    hotel_country = arr[21]
    hotel_market = arr[22]
    is_booking = int(arr[18])
    
    if is_booking == 0:
        continue
        
    if checkInDate != '':
        checkInDateDT = datetime.strptime(checkInDate, "%Y-%m-%d")
        checkInDayOfYear = checkInDateDT.timetuple().tm_yday
        checkInFreqByHotelMarket[(hotel_country, hotel_market)][checkInDayOfYear] += 1
        
    if checkOutDate != '':
        checkOutDateDT = datetime.strptime(checkOutDate, "%Y-%m-%d")
        checkOutDayOfYear = checkOutDateDT.timetuple().tm_yday
        checkOutFreqByHotelMarket[(hotel_country, hotel_market)][checkOutDayOfYear] += 1
    
print('Closing train...')
f.close()
    
print('Opening test...')
f = open("data/test.csv", "r")
f.readline()
total = 0
    
while 1:
    line = f.readline().strip()
    total += 1
        
    if total % 1000000 == 0:
        print('Read {} lines...'.format(total))
    if line == '':
        break

    arr = line.split(",")
    checkInDate = arr[12]
    checkOutDate = arr[13]
    hotel_country = arr[20]
    hotel_market = arr[21]
        
    if checkInDate != '':
        try:
            checkInDateDT = datetime.strptime(checkInDate, "%Y-%m-%d")
            checkInDayOfYear = checkInDateDT.timetuple().tm_yday
            checkInFreqByHotelMarket[(hotel_country, hotel_market)][checkInDayOfYear] += 1
        except:
            pass
        
    if checkOutDate != '':
        try:
            checkOutDateDT = datetime.strptime(checkOutDate, "%Y-%m-%d")
            checkOutDayOfYear = checkOutDateDT.timetuple().tm_yday
            checkOutFreqByHotelMarket[(hotel_country, hotel_market)][checkOutDayOfYear] += 1
        except:
            pass
    
print('Closing test...')
f.close()
```

```python
hotel_market = ('50', '647')
x = checkInFreqByHotelMarket[hotel_market].keys()
y = checkInFreqByHotelMarket[hotel_market].values()
plt.plot(x, y, 'ro')
```

```python
p = np.poly1d(np.polyfit(x, y, 3))
xp = np.linspace(0, 366, 100)
plt.plot(x, y, '.', xp, p(xp), '-', xp, p(xp), '--')
```

```python
checkInDensityByHotelMarket = dict()
checkOutDensityByHotelMarket = dict()
```

```python
for key, value in checkInFreqByHotelMarket.iteritems():
    p = np.poly1d(np.polyfit(value.keys(), value.values(), 3))
    checkInDensityByHotelMarket[key] = p
    
for key, value in checkOutFreqByHotelMarket.iteritems():
    p = np.poly1d(np.polyfit(value.keys(), value.values(), 3))
    checkOutDensityByHotelMarket[key] = p
```

```python
pickle.dump(checkInDensityByHotelMarket, open("data/checkInDensityByHotelMarket.p", "wb"))
pickle.dump(checkOutDensityByHotelMarket, open("data/checkOutDensityByHotelMarket.p", "wb"))
```
